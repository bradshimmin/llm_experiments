```python
from transformers import AutoTokenizer
import transformers
import torch
```

```python

model = "codellama/CodeLlama-7b-Instruct-hf" #"codellama/CodeLlama-7b-hf"

```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
system = "Provide answers in Python. Please comment where appropriate."
user = "Write a function that prints the first 'n' numbers of the fibonacci series."


prompt = f"<s><<SYS>>\n{system}\n<</SYS>>\n\n{user}"
```

```python
sequences = pipeline(
    prompt,
    do_sample=True,
    top_k=10,
    temperature=0.1,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
    add_special_tokens=False
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

```python
# test code

def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a)
        a, b = b, a + b
    return a

fibonacci(10)
```
